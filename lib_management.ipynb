```python

class Library():
    def __init__(self,list):
        self.books_list=list
        self.available_books_list=list
        self.books_lend={}
    def display_available_books(self):
        for book in self.available_books_list:
            print(book)
    def display_all_books(self):
         for book in self.books_list:
            print(book)
    def lend_books(self,name,book):
        if book not in self.books_list:
            print("Incorrect book name.please check the book list")
            return
        if book in self.available_books_list:
            self.books_lend.update({book:name})
            self.available_books_list.remove(book)
            print("you can take the book")
        else:
            print("the book is already taken by"+self.books_lend[book])
    def return_book(self,book):
        if book in self.books_lend:
            del self.books_lend[book]
            self.available_books_list.append(book)
        else:
            print("This book is not borrowed by anyone")

if __name__=="__main__":
    lib=Library(["The life Divine","tamil","english","maths","physics"])
    print("welcome to Library.Enter an option")
    while True:
        print("1.Display available books")
        print("2.Display all books")
        print("3.Borrow a book")
        print("4.Return a book")
        print("5.Quit")
        user_choice=int(input("Enter your choice (1 to 5) "))
        if user_choice==1:
            lib.display_available_books()
            print("___________________________________________________________________")
        elif user_choice==2:
             lib.display_all_books()
             print("___________________________________________________________________")
        elif user_choice==3:
            name=input("enter user name:")
            book=input("enter book name:")
            lib.lend_books(name,book)
            print("___________________________________________________________________")
        elif user_choice==4:
            book=input("enter the book name:")
            lib.return_book(book)
            print("___________________________________________________________________")
        elif user_choice==5:
            break
        else:
            print("Invalid choice..!")
```
